```python
import numpy as np
import pandas as pd

from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
from sklearn.model_selection import train_test_split
```

```python
iris = pd.read_csv('iris.csv')
X = iris.iloc[:, :-1].values
y = iris.iloc[:, -1].values

model = KMeans()
visualizer = KElbowVisualizer(model, k = (2, 15))
visualizer.fit(X)
visualizer.poof()
```

Based on this Elbow Plot for the Distortion Score for K Means Clustering on the iris dataset, the optimal number of clusters for this dataset is 3. This corroborates known fact, since everyone knows the iris dataset contains data on 3 specific flowers: `setosa`, `versicolor`, and `virginica`.

```python
def calculate_sse(data, centroid):
    sse = 0
    for row in data:
        sse += (sum(row - centroid) ** 2)
    return sse

```
