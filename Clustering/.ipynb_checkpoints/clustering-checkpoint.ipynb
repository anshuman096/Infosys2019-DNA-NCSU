```python
from sklearn.datasets import make_moons, make_circles, make_blobs
from sklearn.datasets import make_gaussian_quantiles

from sklearn.cluster import KMeans, DBSCAN, SpectralClustering, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np

% matplotlib inline
```

# Creating datasets

```python
X, y = make_gaussian_quantiles(n_features=2, n_classes=3, n_samples=500, random_state=1)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=25, edgecolor='k')
plt.show()
```

# Clustering
## K-Means

```python
X, y = make_blobs(n_samples=1500, centers=4, cluster_std=0.60, random_state=1)
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.show()
```

```python
km = KMeans(n_clusters=2, random_state=0).fit(X)
plt.scatter(X[:, 0], X[:, 1], s=50, c=km.labels_, edgecolor='k')
plt.show()
```

```python
# Sum of square distances of the samples
km.inertia_
```

```python
k_list = range(2,11)
ssd = [KMeans(n_clusters=k, random_state=0).fit(X).inertia_ for k in k_list]
```

```python
plt.plot(k_list, ssd, 'X-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
```

```python
km = KMeans(n_clusters=4, random_state=0).fit(X)
plt.scatter(X[:, 0], X[:, 1], s=50, c=km.labels_, edgecolor='k')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], c='g')
```

```python
km.cluster_centers_[:,0]
```

# Spectral Clustering
In practice Spectral Clustering is very useful when the structure of the individual clusters is highly non-convex or more generally when a measure of the center and spread of the cluster is not a suitable description of the complete cluster. For instance when clusters are nested circles on the 2D plan.

```python
X, y = make_blobs(n_samples=1500, centers=4, cluster_std=0.60, random_state=1)
plt.scatter(X[:, 0], X[:, 1], s=50)
```

```python
spec = SpectralClustering(n_clusters=4, assign_labels="discretize", random_state=1).fit(X)
plt.scatter(X[:, 0], X[:, 1], s=50, c=spec.labels_, edgecolor='k')
```

## moons

```python
X, y = make_moons(1500, noise=.05, random_state=1)
plt.scatter(X[:, 0], X[:, 1], s=50)
```

```python
spec = SpectralClustering(affinity='rbf', assign_labels="discretize",eigen_solver='arpack', random_state=1).fit(X)
plt.scatter(X[:, 0], X[:, 1], s=50, c=spec.labels_, edgecolor='k')
```

```python
spec = SpectralClustering(affinity="nearest_neighbors", n_clusters=2, eigen_solver='arpack', random_state=1).fit(X)
plt.scatter(X[:, 0], X[:, 1], s=50, c=spec.labels_, edgecolor='k')
```

## gaussian

```python
X, y = make_gaussian_quantiles(n_features=2, n_classes=3,n_samples=1000, random_state=1)
plt.scatter(X[:, 0], X[:, 1], marker='o')
```

```python
spec = SpectralClustering(affinity="rbf", n_clusters=3, eigen_solver='arpack', random_state=1).fit(X)
plt.scatter(X[:, 0], X[:, 1], s=50, c=spec.labels_, edgecolor='k')
```

```python
spec = SpectralClustering(affinity="nearest_neighbors", n_clusters=3, eigen_solver='arpack', random_state=1).fit(X)
plt.scatter(X[:, 0], X[:, 1], s=50, c=spec.labels_, edgecolor='k')
```

## circle

```python
X, y = make_circles(n_samples=1500, noise=0.05,factor=0.5, random_state=1)
plt.scatter(X[:, 0], X[:, 1], marker='o')
```

# Agglomerative Clustering
Recursively merges the pair of clusters that minimally increases a given linkage distance.

```python
X, y = make_blobs(n_samples=1500, centers=4, cluster_std=0.60, random_state=1)
plt.scatter(X[:, 0], X[:, 1], s=50)
```

```python
plt.figure(figsize=(10,7))
plt.title("Blobs Dendrogram")
dend = dendrogram(linkage(X, method='ward'))
```

```python
agg = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward').fit(X)
plt.scatter(X[:, 0], X[:, 1], s=50, c=agg.labels_, edgecolor='k')
```

## moons
